#**Creating a Conditional Column from 2 Choices: np.where()**

```python
import numpy as np
```

```python
# create an array based on price tiers
np.where(df_laptops['Price_euros'] >2000, 'Expensive','Cheap')
```

```python
# set it to a new column
df_laptops['Price_tier'] = np.where(df_laptops['Price_euros']>2000, 'Expensive', 'Cheap')
```

```python
# show first 5 rows
df_laptops.head()
```

```python
# count values in price_tier column
df_laptops.value_counts('Price_tier')
```

```python
# create an array based on screen size (screen size greather than 15 is 'Big' and less or equal to 15 is 'Small')
np.where(df_laptops['Inches']>15,'Big','Small')
```

```python
# set it to a new column named 'Scren_size'
df_laptops['screen_size'] = np.where(df_laptops['Inches']>15,'Big','Small')
```

```python
# show first 5 rows
df_laptops.head()
```

```python
# count values in 'Scren_size' column
df_laptops.value_counts('screen_size')
```
